```python
#This assignment is for visualization using matplotlib:
#Data to use:
#url= https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv '
#titanic = pd.read_csv(url)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url= "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
```

```python
#Charts to plot:

#1. Create a pie chart presenting the male/female proportion

#Plotting the piechart
plt.pie(titanic.sex.value_counts(), # List of total male and female counts on titanic
        labels = ["Males","Females"], labeldistance=0.6, # Labels and their distances from the center of the piechart
        colors=['deepskyblue',"hotpink"], # Colors of the pies in the pie-chart
        autopct='%2.2f%%',  pctdistance=0.4, # Show percentages-  Decimals and their distances from the center of the piechart
        startangle= 130, radius=1,explode=[0,0.05]) # Pie chart appearance - Startangle, radius and pie distance from the center
```

```python
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
fare_male= titanic.fare[titanic.sex=="male"]
fare_female= titanic.fare[titanic.sex=="female"]
ages_male= titanic.age[titanic.sex=="male"]
ages_female= titanic.age[titanic.sex=="female"]

fig = plt.figure(figsize=(18,5))
plt.title("Scatterplot of Fare paid vs Age on Titanic\n")
plt.xlabel("Age",fontsize=18)
plt.ylabel("Fare Paid",fontsize=18)

plt.scatter(ages_male, fare_male, c = 'deepskyblue',label = "Males")
plt.scatter(ages_female, fare_female, c = 'hotpink',label = "Females")

plt.legend()
plt.show()
```
